```python
class Pagamento():
    def __init__(self, valor, moeda):
        self.valor = valor
        self.moeda = moeda 

    def processar_pagamento(self):
        return f"A processar pagamento de {self.valor} {self.moeda}"
```

```python
class PagamentoOnline():
    def __init__(self, valor, moeda, email):
        self.valor = valor
        self.moeda = moeda
        self.email = email

    def processar_pagamento2(self):
        return f"A processar pagamento online de {self.valor} {self.moeda} para o email {self.email}"
```

```python
class CartaoCredito(PagamentoOnline, Pagamento):
    def __init__(self, valor, moeda, email=None):
        Pagamento.__init__(self, valor, moeda)
        PagamentoOnline.__init__(self, valor, moeda, email)
    
    def processar_pagamento2(self):
        return f"{super().processar_pagamento2()}"

    def processar_pagamento(self):
        return f"{super().processar_pagamento()}"
    
```

```python
class TransferenciaBancaria(Pagamento, PagamentoOnline):
     def __init__(self, valor, moeda, email=None):
        Pagamento.__init__(self, valor, moeda)
        PagamentoOnline.__init__(self, valor, moeda, email)
    
     def processar_pagamento(self):
        return f"{super().processar_pagamento()}"
    
     def processar_pagamento2(self):
        return f"{super().processar_pagamento2()}"
```

```python

cartao = CartaoCredito("120.02", "dolar","eduardo@gmail.com")
print(cartao.processar_pagamento2())
transferencia = TransferenciaBancaria("202.10", "euro")
print(transferencia.processar_pagamento())
```

```python
class Despesa:
    def __init__(self, id, descricao, valor):
        self.dados = {
            "id": id,
            "descricao": descricao,
            "valor": valor
        }
    
    def mostrar_informacao(self):
        for chave, valor in self.dados.items():
            print(f"{chave.capitalize()}: {valor}")
    
    def calcular_total_anual(self):
        pass
```

```python
class DespesaFixa(Despesa):
    def __init__(self, id, descricao, valor, frequencia):
        super().__init__(id, descricao, valor)
        self.dados["frequencia"] = frequencia
    
    def mostrar_informacao(self):
        super().mostrar_informacao()
        print(f"Frequência: {self.dados['frequencia']}")
    
    def calcular_total_anual(self):
        frequencia = self.dados['frequencia']
        valor = self.dados['valor']
        
        if frequencia.lower() == "mensal":
            total_anual = valor * 12  
        elif frequencia.lower() == "semanal":
            total_anual = valor * 52  
        else:
            total_anual = 0  

        return total_anual
```

```python
class DespesaVariavel(Despesa):
    def __init__(self, id, descricao, valor, categoria):
        super().__init__(id, descricao, valor)
        self.dados["categoria"] = categoria
    
    def mostrar_informacao(self):
        super().mostrar_informacao()
        print(f"categoria:{self.dados['categoria']}")
    
    def calcular_total_anual(self):
        valor_mensal = self.dados["valor"]
        total_anual = valor_mensal * 12
        return total_anual
```

```python
class  DespesaOcasional(Despesa):
    def __init__(self, id, descricao, valor, data):
        super().__init__(id, descricao, valor)
        self.dados["data"] = data
    
    def mostrar_informacao(self):
        super().mostrar_informacao()
        print(f"data:{self.dados['data']}")
    
    def calcular_total_anual(self):
        return self.dados["valor"]
    
```

```python
despesa_fixa = DespesaFixa(1, "Aluguel de apartamento", 1200.00, "Mensal")

# Despesa Variável
despesa_variavel = DespesaVariavel(2, "Gás de cozinha", 50.00, "Mensal")

# Despesa Ocasional
despesa_ocasional = DespesaOcasional(3, "Reparos no carro", 800.00, "2024-05-20")

# Exibindo informações e cálculos
print("\nInformações da Despesa Fixa:")
despesa_fixa.mostrar_informacao()
print(f"Total Anual: {despesa_fixa.calcular_total_anual()}")

print("\nInformações da Despesa Variável:")
despesa_variavel.mostrar_informacao()
print(f"Total Anual: {despesa_variavel.calcular_total_anual()}")

print("\nInformações da Despesa Ocasional:")
despesa_ocasional.mostrar_informacao()
print(f"Total Anual: {despesa_ocasional.calcular_total_anual()}")
```

```python
def ler_numero():
    while True:
        numero = input("Digite um número>")
        try:
            return float(numero)  
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

def maior(a, b, c):
    return max(a, b, c)


print("Vamos encontrar o maior entre três números.")
num1 = ler_numero()
num2 = ler_numero()
num3 = ler_numero()


maior_numero = maior(num1, num2, num3)
print(f"O maior número entre {num1}, {num2} e {num3} é: {maior_numero}")
```

```python
class CalcularIMC:
    def peso(self):
        while True:
            peso = input("Digite o seu peso (kg)>")
            try:
                return float(peso)
            except ValueError:
                print("Entrada inválida. Por favor, digite um número.")

    def altura(self):
        while True:
            altura = input("Digite a sua altura (m)>")
            try:
                return float(altura)
            except ValueError:
                print("Entrada inválida. Por favor, digite um número.")

    def imc(self):
        peso = self.peso() 
        altura = self.altura() 
        imc = peso / (altura ** 2) 

        
        if imc < 18.5:
            print(f"{imc:.2f}, Você está abaixo do peso!")
        elif 18.5 <= imc < 25:
            print(f"{imc:.2f}, Seu peso está normal!")
        elif 25 <= imc < 30:
            print(f"{imc:.2f}, Você está acima do peso!")
        else:
            print(f"{imc:.2f}, Você está obeso!")


calculo_imc = CalcularIMC()
calculo_imc.imc()
```

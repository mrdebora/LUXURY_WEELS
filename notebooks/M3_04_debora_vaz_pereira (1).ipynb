```python
class ponto():
    x = 0
    y = 0
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"{self.x},{self.y}"
    
    def quadrante(self):
        if self.x > 0 and self.y > 0:
            return "primeiro quadrante"
        elif self.x < 0 and self.y > 0:
            return "segundo quadrante"
        elif self.x < 0 and self.y < 0:
            return "terceiro quadrante"
        elif self.x > 0 and self.y < 0:
            return "quarto quadrante"
        elif self.x == 0 and self.y == 0:
            return "origem"
        elif self.x == 0:
            return "eixo y"
        elif self.y == 0:
            return "eixo x"
    
    def vetor(self, outro_ponto):
        vetor_x = outro_ponto.x - self.x
        vetor_y = outro_ponto.y - self.y
        return ponto(vetor_x, vetor_y)
```

```python
A = ponto(2, 3)
B = ponto(5, 5)
C = ponto(-3, -1)
D = ponto(0, 0)

print(f"Ponto A: {A} está no {A.quadrante()}")
print(f"Ponto C: {C} está no {C.quadrante()}")
print(f"Ponto D: {D} está no {D.quadrante()}")

AB = A.vetor(B)
BA = B.vetor(A)

print(f"Vetor AB: {AB}")
print(f"Vetor BA: {BA}")
```

```python
class Retângulo():
    def __init__(self, ponto_inicial, ponto_final):
        self.ponto_inicial = ponto_inicial
        self.ponto_final = ponto_final

    def base(self):
        return abs(self.ponto_final.x - self.ponto_inicial.x)

    def altura(self):
        return abs(self.ponto_final.y - self.ponto_inicial.y)

    def area(self):
        return self.base() * self.altura()

    def __str__(self):
        return f"Retângulo[Inicial: {self.ponto_inicial}, Final: {self.ponto_final}]"
 
```

```python
class Veiculo():
    def __init__(self, cor, rodas):
        self.cor = cor
        self.rodas = rodas 

    def __str__(self):
        return f"veiculo:Nº de rodas:{self.rodas}, cor:{self.cor}"

class Carro(Veiculo):
    def __init__(self, cor, rodas,velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada

    def __str__(self):
        return f"{super().__str__()},Velocidade:{self.velocidade},Cilindrada:{self.cilindrada}"

class Bicicleta(Veiculo):
    def __init__(self, cor, rodas):
        super().__init__(cor, rodas)

    def __str__(self):
        return f"{super().__str__()}"

class Moto(Veiculo):
    def __init__(self, cor, rodas, velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada

    def __str__(self):
        return f"{super().__str__()},Velocidade:{self.velocidade},Cilindrada:{self.cilindrada}"

class Caminhao(Veiculo):
    def __init__(self, cor, rodas, velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada

    def __str__(self):
        return f"{super().__str__()},Velocidade:{self.velocidade},Cilindrada:{self.cilindrada}"

class Concessionario():
    def catalogar(self, lista_veiculos):
        for veiculo in lista_veiculos:
            print(f"Classe: {type(veiculo).__name__}, Atributos: {veiculo}")

    def filtrar(self, lista_veiculos, tipo_veiculo):
        lista_filtrada = [veiculo for veiculo in lista_veiculos if isinstance(veiculo, tipo_veiculo)]
        return lista_filtrada

carro1 = Carro("Vermelho", 4, 180, 1600)
carro2 = Carro("Azul", 4, 200, 2000)
bicicleta1 = Bicicleta("Verde", 2)
bicicleta2 = Bicicleta("Amarelo", 2)
moto1 = Moto("Preto", 2, 150, 600)
moto2 = Moto("Branco", 2, 180, 800)
caminhao1 = Caminhao("Cinza", 6, 120, 8000)
caminhao2 = Caminhao("Azul", 6, 110, 9000)


lista_veiculos = [carro1, carro2, bicicleta1, bicicleta2, moto1, moto2, caminhao1, caminhao2]

concessionario = Concessionario()

print("Catalogação dos veículos:")
concessionario.catalogar(lista_veiculos)

tipo_desejado = Carro 
veiculos_filtrados = concessionario.filtrar(lista_veiculos, tipo_desejado)

print("\nVeículos filtrados (Carros):")
for veiculo in veiculos_filtrados:
    print(veiculo)
```

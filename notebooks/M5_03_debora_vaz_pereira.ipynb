```python
import json
```

```python
dados = {
    "frutaria": [
        {
            "fruta": [
                {
                    "nome": "abacate",
                    "quantidade" : 10
                },
                {
                    "nome": "banana",
                    "quantidade" : 15
                },
                {
                    "nome": "laranja",
                    "quantidade" : 18
                },
                {
                    "nome": "morango",
                    "quantidade" : 19
                }
            ]
        },
        {
            "verdura": [
                {
                    "nome":"brócolis",
                    "quantidade": 13
                },
                {
                    "nome":"cenoura",
                    "quantidade": 21
                },
                {
                    "nome":"batata",
                    "quantidade": 26
                },
                {
                    "nome":"repolho",
                    "quantidade": 30
                }
            ]
        }
    ]
}

with open("frutaria.json", "w") as f:
    json.dump(dados, f, indent=4)
```

```python
with open("frutaria.json", "r") as f:
    ficheiro = json.load(f)


frutas = ficheiro["frutaria"][0]["fruta"]
verduras = ficheiro["frutaria"][1]["verdura"]

for fruta in frutas:
    print(f'{fruta["nome"]}: {fruta["quantidade"]}')

for verdura in verduras:
    print(f'{verdura["nome"]}: {verdura["quantidade"]}')
```

```python
dados = {
'marcadores': [{'latitude': 40.416875,
   'longitude': -3.703308,
   'city': 'madrid',
   'description': 'puerta del sol'},
  {'latitude':  40.417438,
   'longitude':  -3.693363,
   'city': 'madrid',
   'description': 'paseo del prado'},
  {'latitude': 40.407015,
   'longitude': -3.691163,
   'city': 'madrid',
   'description': 'estación de atocha'}]
}
```

```python
dados_json = [
    {"nome":"João", "idade": 28, "interesses": ["musica", "futebol", "cinema"]},
    {"nome":"maria", "idade": 34, "interesses": ["literatura", "viajar", "pintura"]},
    {"nome": "ana", "idade": 22, "interesses": ["fotografia", "moda", "tecnologia"]}
]
```

```python
dados_convertidos = []
for pessoa in dados_json:
    novo = {
        "Nome": pessoa["nome"].capitalize(),
        "Idade": pessoa["idade"],
        "Interesse 1": pessoa["interesses"][0],
        "Interesse 2": pessoa["interesses"][1],
        "Interesse 3": pessoa["interesses"][2]
    }
    dados_convertidos.append(novo)
```

```python
import csv

campos = ["Nome", "Idade", "Interesse 1", "Interesse 2", "Interesse 3"]


with open("usuarios.csv", "w", newline='', encoding='utf-8') as arquivo_csv:
    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos)
    escritor.writeheader()
    escritor.writerows(dados_convertidos)


with open("usuarios.csv", "r", encoding="utf-8") as f:
    conteudo = f.read()
    print(conteudo)
```

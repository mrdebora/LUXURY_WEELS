```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()

cursor.execute(''' 
SELECT cargo, MIN(salario) AS salario_minimo, 
Max(salario) AS salario_maximo, AVG(salario) AS salario_medio 
FROM Funcionarios GROUP BY cargo ORDER BY salario_medio DESC''')

disparidades = cursor.fetchall()
print("\nDisparidades salariais entre cargos:")
for linha in disparidades:
    print(linha)

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()


#AVG(salario) faz o calculo do salario medio e guardar em Funcionarios ele agrupa os cargos com GROUP e lista os salarios medio em ordem descente com DESC e aparece apenas um por conta do LIMIT 1
cursor.execute('''
SELECT cargo, AVG(salario) as salario_medio
FROM Funcionarios GROUP BY cargo ORDER BY salario_medio DESC
LIMIT 1
''')
maior_salario_medio = cursor.fetchone()
print("\nCargo com maior salário médio>", maior_salario_medio)

cursor.execute('''
SELECT cargo, AVG(salario) as salario_medio
FROM Funcionarios GROUP BY cargo ORDER BY salario_medio ASC
LIMIT 1
''')
menor_salario_medio = cursor.fetchone()
print("\nCargo com menor salário médio>", menor_salario_medio)

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()

#CONT(*) conta quantos funcionarios tem em cada cargo e coloca em quantidade_funcionarios, avg(salario) calcula a media e depois e ordenado em ordem descente...
cursor.execute('''
SELECT cargo, COUNT(*) AS quantidade_funcionarios, AVG(salario) AS salario_medio
FROM Funcionarios GROUP BY cargo ORDER BY quantidade_funcionarios DESC
''')

quantidade_por_cargo = cursor.fetchall()
print("\nQuantidade de funcionários e salário médio por cargo:")
for linha in quantidade_por_cargo:
    print(linha)

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE Projetos (
    id_projeto INTEGER PRIMARY KEY,
    nome_projeto TEXT,
    funcionario_id INTEGER,
    FOREIGN KEY (funcionario_id) REFERENCES Funcionarios(id))''')


conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()

cursor.execute('''
SELECT Funcionarios.nome AS Nome_Funcionario, 
Projetos.nome_projeto AS Nome_Projeto FROM Funcionarios
JOIN Projetos ON Funcionarios.id = Projetos.funcionario_id;
''')

resultados = cursor.fetchall()

for linha in resultados:
    print(f"Funcionário: {linha[0]} - Projeto: {linha[1]}")

conn.close()
```

```python
texto = "Olá Mundo"
texto_alinhado_direita = texto.rjust(20)
print(f"'{texto_alinhado_direita}'")
```

```python
texto = "Olá Mundo"
texto_truncado = texto[:4]
print(texto_truncado) 
```

```python
texto = "Olá Mundo"

texto_truncado = texto[:2] 
texto_alinhado_centro = texto_truncado.center(20)
print(f"'{texto_alinhado_centro}'") 
```

```python
numero = 150
numero_formatado = f"{numero:-05}"
print(numero_formatado)  
```

```python
numero = 7887
numero_formatado = str(numero).rjust(7)
print(f"'{numero_formatado}'")  # Saída: '   7887'
```

```python
numero = 20.02
numero_formatado = f"{numero:6.3f}"
print(numero_formatado)
```

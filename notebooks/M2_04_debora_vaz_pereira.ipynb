```python
class Aluno():
    def __init__(self, nome, nota):
        self.__nome = nome
        self.__nota = float(nota)
        print('O aluno foi criado!')
        
    def __str__(self):
        return f"Aluno:{self.__nome}, Nota:{self.__nota}"

    def qualificação(self):
        if self.__nota >= 6:
            print(f"O aluno {self.__nome} passou com a nota {self.__nota}.")
        else:
            print(f"O aluno {self.__nome} reprovou com a nota {self.__nota}.")
 

   
    def get_nome(self):
        return self.__nome 

    def get_nota(self):
        return self.__nota
    
    def set_nome(self, nome):
        self.__nome = nome
    
    def set_nota(self, nota):
        try:
            self.__nota = float(nota) 
        except ValueError:
            print("Nota inválida! Deve ser um número.")

        
 
```

```python
aluno1 = Aluno("debora", "10")
aluno2 = Aluno("bia", "5")

print(aluno1)
print(aluno2)
```

```python
aluno1.qualificação()
aluno2.qualificação()



aluno1.set_nome("Vitor")
aluno1.set_nota(9)
print(aluno1)

```

```python
class Produto():
    def __init__(self, codigo, nome, preco, tipo):
        self.__codigo = codigo
        self.__nome = nome
        self.__preco = preco
        self.__tipo = tipo
        print("O produto foi criado com sucesso")

    def __str__(self):
        return f"Produto {self.__codigo}: {self.__nome}, Preço: {self.__preco}, Tipo: {self.__tipo}"

    def  calcular_total(self, unidades): 
        return self.__preco * unidades

    def get_codigo(self):
        return self.__codigo

    def get_nome(self):
        return self.__nome

    def get_preco(self):
        return self.__preco

    def get_tipo(self):
        return self.__tipo

    def set_codigo(self, codigo):
        self.__codigo = codigo

    def set_nome(self, nome):
        self.__nome = nome

    def set_preco(self, preco):
        self.__preco = preco

    def set_tipo(self, tipo):
        self.__tipo = tipo
```

```python
produto1 = Produto(codigo=101, nome="Notebook", preco=2500.00, tipo="Eletrônico")
produto2 = Produto(codigo=102, nome="Camiseta", preco=49.90, tipo="Vestuário")
produto3 = Produto(codigo=103, nome="Mesa", preco=399.99, tipo="Móvel")
produto4 = Produto(codigo=104, nome="Smartphone", preco=1500.00, tipo="Eletrônico")
produto5 = Produto(codigo=105, nome="Livros", preco=99.90, tipo="Material Didático")


print(produto1)
print(produto2)
print(produto3)
print(produto4)
print(produto5)

unidades = 3  
total_produto1 = produto1.calcular_total(unidades)
total_produto2 = produto2.calcular_total(unidades)

print(f"\nTotal para {unidades} unidades do {produto1.get_nome()}: R${total_produto1:.2f}")
print(f"Total para {unidades} unidades do {produto2.get_nome()}: R${total_produto2:.2f}")
```

```python
class Pedido:
    def __init__(self, produtos, quantidades):
        self.produtos = produtos 
        self.quantidades = quantidades 

    def total_pedido(self):
        total = 0
        for produto, quantidade in zip(self.produtos, self.quantidades):
            total += produto.get_preco() * quantidade
        return total

    def mostrar_pedido(self):
        for produto, quantidade in zip(self.produtos, self.quantidades):
            print(f"Produto: {produto.get_nome()}, Quantidade: {quantidade}")

```

```python
produto1 = Produto(101, "Notebook", 2500.00, "Eletrônico")
produto2 = Produto(102, "Camiseta", 49.90, "Vestuário")
produto3 = Produto(103, "Mesa", 399.99, "Móvel")
produto4 = Produto(104, "Smartphone", 1500.00, "Eletrônico")
produto5 = Produto(105, "Livros", 99.90, "Material Didático")
produto6 = Produto(106, "Caderno", 19.90, "Material Didático")


produtos_pedido1 = [produto1, produto2, produto3]  
quantidades_pedido1 = [1, 2, 1] 

pedido1 = Pedido(produtos_pedido1, quantidades_pedido1)


print("Pedido 1:")
pedido1.mostrar_pedido()
total1 = pedido1.total_pedido()
print(f"Total do Pedido 1: R${total1:.2f}\n")


produtos_pedido2 = [produto4, produto5, produto6]  
quantidades_pedido2 = [1, 3, 5] 


pedido2 = Pedido(produtos_pedido2, quantidades_pedido2)

print("Pedido 2:")
pedido2.mostrar_pedido()
total2 = pedido2.total_pedido()
print(f"Total do Pedido 2: R${total2:.2f}")
```

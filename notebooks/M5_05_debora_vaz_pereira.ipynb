```python
import numpy as np

array_10_100 = np.arange(10, 101)
print("Array de 10 a 100>")
print(array_10_100)

array_invertido = array_10_100[::-1]
print("\nArray invertido>")
print(array_invertido)

array_3x3 = np.arange(9).reshape(3, 3)
print("\nArray 3x3 com valores de 0 a 8>")
print(array_3x3)

identidade_6x6 = np.eye(6)
print("\nArray identidade 6x6>")
print(identidade_6x6)

array_random_3x3x3 = np.random.rand(3, 3, 3)
print("\nArray aleatório 3x3x3>")
print(array_random_3x3x3)

A = np.random.rand(3, 3)
B = np.random.rand(3, 3)

iguais_elemento_a_elemento = np.array_equal(A, B)
iguais_proximos = np.allclose(A, B)

print("\nArray A>")
print(A)
print("\nArray B>")
print(B)
print("\nArray A e B sao exatamente iguais?>", iguais_elemento_a_elemento)
print("Array A e B sao aproximadamente iguais?>", iguais_proximos)
```

```python
import numpy as np

dolares = np.random.randint(1, 1001, size=100)
print("Array em dólares>")
print(dolares)

taxa_conversao = 0.874413

euros = dolares * taxa_conversao

euros_arredondado = np.round(euros, 2)

print("\nArray convertido para euros>")
print(euros_arredondado)
```

```python
import pandas as pd

df_titulos = pd.read_csv('05_05_imdb_titulos.csv')
print("Primeiros 5 registros de títulos>")
print(df_titulos.head())

df_elencos = pd.read_csv('05_05_imdb_elenco.csv')
print("\nPrimeiros 5 registros de elencos>")
print(df_elencos.head())

print("\nNumero de registros do dataframe de titulos>", len(df_titulos))

print("\nNumero de registros do dataframe de elencos>", len(df_elencos))

df_titulos_mais_antigos = df_titulos.sort_values(by='year').head(5)
print("\n5 titulos mais antigos")
print(df_titulos_mais_antigos[['title', 'year']])

dracula_filmes = df_titulos[df_titulos['title'].str.contains('dracula', case=False, na=False)]
print("\nFilmes que contem 'Drácula' no título>")
print(dracula_filmes[['title', 'year']])

print("\nNumeros de filmes que contem 'Drácula' no titulo>", len(dracula_filmes))

titulos_mais_comuns = df_titulos['title'].value_counts().head(10)
print("\n10 Títulos mais comuns:")
print(titulos_mais_comuns)

romeo_juliet_filmes = df_titulos[df_titulos['title'].str.contains('romeo', case=False, na=False) & 
                                  df_titulos['title'].str.contains('juliet', case=False, na=False)]

print("\nFilmes com 'Romeo' e 'Juliet' no título:")
print(romeo_juliet_filmes.sort_values(by='year')[['title', 'year']])
```

```python
exorcist_filmes = df_titulos[df_titulos['title'].str.contains('Exorcist', case=False, na=False)].sort_values('year')
print(exorcist_filmes)
```

```python
filmes_1950 = df_titulos[df_titulos['year'] == 1950]
print(len(filmes_1950))
```

```python
filmes_anos_50 = df_titulos[(df_titulos['year'] >= 1950) & (df_titulos['year'] <= 1959)]
print(len(filmes_anos_50))
```

```python
godfather_filmes = df_elencos[df_elencos['title'].str.contains('The Godfather', case=False, na=False)]
papel_contagem = godfather_filmes.groupby('character').size().reset_index(name='quantidade')
papel_contagem = papel_contagem.sort_values('quantidade', ascending=False)
print(papel_contagem)
```

```python
dracula_1958 = df_elencos[(df_elencos['title'].str.contains('dracula', case=False, na=False)) & (df_elencos['year'] == 1958)]
dracula_1958_sorted = dracula_1958.sort_values('n')
print(dracula_1958_sorted)
```

```python
bruce_wayne_roles = df_elencos[(df_elencos['character'].str.contains('Bruce wayne', case=False , na=False)) & (df_elencos['n'] == 1)]
print(len(bruce_wayne_roles))
```

```python
deniro_roles = df_elencos[df_elencos['name'].str.contains('Robert De Niro', case=False)]
print(len(deniro_roles))
```

```python
heston_60s = df_elencos[
    (df_elencos['name'].str.contains('Charlton Heston', case=False)) &
    (df_elencos['n'] == 1) &
    (df_elencos['year'].between(1960, 1969))
]
heston_60s_sorted = heston_60s.sort_values('year', ascending=False)
print(heston_60s_sorted)
```

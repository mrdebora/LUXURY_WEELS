```python
class Evento:
    def __init__(self, titulo, data_hora_inicio, data_hora_fim):
        self.titulo = titulo 
        self.data_hora_inicio = data_hora_inicio 
        self.data_hora_fim = data_hora_fim
        #         já converte na data e hora e nomeando ^^

    def __str__(self):
       return f"{self.titulo}: {self.data_hora_inicio} até {self.data_hora_fim}"
class Agenda:
    def __init__(self):
        self.lista_eventos = []

    def adicionar_evento(self,Evento):
        self.lista_eventos.append(Evento)

    def listar_evento(self):
        for evento in self.lista_eventos:
            print(evento)

    def proximo_evento(self):
        data_atual = datetime.now()
        
    def eventos_do_dia(self, data):
        # Converte a data recebida para o formato correto(.date (ano/mes/dia))
        data_procurada = datetime.strptime(data, "%Y-%m-%d").date()
        eventos_no_dia = [                                                                                                                                                             
            evento for evento in self.lista_eventos if evento.data_hora_inicio.date() == data_procurada
        ]
        # Evento é um objeto temporio que servira para corre a lista e verificar somente a data com o evento.data_hora_inicio.date() e vai comparar as datas 
        if not eventos_no_dia:
            print(f"Não há eventos no dia {data}.")
            # se a evento.data_hora_inicio.date() não achar nada igual a data_procurada percorre isto 
        else:
            print(f"Eventos no dia {data}:")
            # foi criado um nova lista eventos_no_dia onde se tiver mais de um evento no dia vai imprimir todos
            for evento in eventos_no_dia:
                print(f"- {evento.titulo}: {evento.data_hora_inicio} até {evento.data_hora_fim}")
        


```

```python
agenda = Agenda()


evento1 = Evento("Reunião de equipe", datetime(2025, 1, 25, 10, 0), datetime(2025, 1, 25, 12, 0))
evento2 = Evento("Almoço com cliente", datetime(2025, 1, 26, 13, 0), datetime(2025, 1, 26, 14, 0))
evento3 = Evento("Treinamento", datetime(2025, 1, 27, 9, 0), datetime(2025, 1, 27, 11, 0))
evento4 = Evento("Planejamento estratégico", datetime(2025, 1, 28, 15, 0), datetime(2025, 1, 28, 17, 0))
evento5 = Evento("Revisão de projeto", datetime(2025, 1, 25, 14, 0), datetime(2025, 1, 25, 16, 0))


agenda.adicionar_evento(evento1)
agenda.adicionar_evento(evento2)
agenda.adicionar_evento(evento3)
agenda.adicionar_evento(evento4)
agenda.adicionar_evento(evento5)


agenda.listar_evento()
print()


agenda.proximo_evento()
print()


data_procurada = datetime(2025, 1, 25).date()
agenda.eventos_do_dia(data_procurada)
```

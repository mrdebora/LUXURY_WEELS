```python
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
```

```python
print("\nEscolha uma das opções:")
print("1) Mostrar a soma dos dois números")
print("2) Mostrar o resto dos dois números (o primeiro menos o segundo)")
print("3) Mostrar a multiplicação dos dois números")
```

```python
opcao = input("\nDigite o número da opção escolhida: ")
```

```python
if opcao == "1":
    soma = numero1 + numero2
    print(f"A soma dos dois números é: {soma}")
elif opcao == "2":
    resto = numero1 - numero2
    print(f"O resto da subtração dos dois números é: {resto}")
elif opcao == "3":
    multiplicacao = numero1 * numero2
    print(f"A multiplicação dos dois números é: {multiplicacao}")
else:
    print("Opção escolhida não é correta.")
```

```python
#exercicio 6
```

```python
numeros_validos = [2, 4, 5, 7]


while True:
    numero = int(input("Digite um número inteiro de 0 a 9: "))
    
    if 0 <= numero <= 9:
        if numero in numeros_validos:
            print(f"O número {numero} está na lista.")
            break  
        else:
            print(f"O número {numero} não está na lista. Tente novamente.")
    else:
        print("Número fora do intervalo de 0 a 9. Tente novamente.")
```

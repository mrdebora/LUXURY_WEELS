```python
class ValidadorPalavraPasse():
    def __init__(self, senha):
        self.senha = senha

    def verificar_letras(self):
        return self.senha.isalpha()
    
    def verificar_digitos(self):
        return self.senha.isdigit()

    def verificar_letras_e_digitos(self):
        tem_letras = any(char.isalpha() for char in self.senha)
        tem_digitos = any(char.isdigit() for char in self.senha)
        return tem_letras and tem_digitos

    def verificar_comprimento(self):
        return len(self.senha) >= 8

    def validar(self):
        print("Verificação de senha:")
        print(f"• Apenas letras: {self.verificar_letras()}")
        print(f"• Apenas dígitos: {self.verificar_digitos()}")
        print(f"• Contém letras e dígitos: {self.verificar_letras_e_digitos()}")
        print(f"• Comprimento mínimo (8 caracteres): {self.verificar_comprimento()}")


```

```python
palavra_passe = "Exemplo123"
validador = ValidadorPalavraPasse(palavra_passe)

validador.validar()
```

```python
class ExtratorTexto:
    def __init__(self, texto, palavra):
        self.texto = texto
        self.palavra = palavra
    
    def pesquisar_e_extrair(self):
        posicao = self.texto.find(self.palavra)

        if posicao == -1:
            print("palavra não encontrada!")
        else:
            final = len(self.texto)
            for pontuacao in ['.','!','?']:
                fim_frase = self.texto.find(pontuacao, posicao)
                if fim_frase != -1:
                    final = min(final, fim_frase + 1)
        trecho = self.texto[posicao:final]
        print(f"{trecho}")
        
```

```python
texto_exemplo = "Python é uma linguagem de programação muito popular. Ela é usada em ciência de dados!"
palavra_pesquisar = "linguagem"

extrator = ExtratorTexto(texto_exemplo, palavra_pesquisar)
extrator.pesquisar_e_extrair()
```

```python
class FraseAnalise:
    def __init__(self):
        self.frase = ""
        self.lista_palavras = []
        self.palavra_busca = ""

    def solicitar_entrada(self):
        # Solicita ao utilizador que insira uma frase e uma palavra para buscar
        self.frase = input("Digite uma frase: ")
        self.palavra_busca = input("Digite uma palavra para buscar na frase: ")

    def processar_frase(self):
        # Divide a frase em palavras e armazena numa lista
        self.lista_palavras = self.frase.split()
        print("\nLista de palavras extraídas da frase:")
        print(self.lista_palavras)

    def contar_ocorrencias(self):
        # Conta e mostra o número de vezes que a palavra aparece na lista
        contagem = self.lista_palavras.count(self.palavra_busca)
        print(f"\nA palavra '{self.palavra_busca}' aparece {contagem} vez(es) na frase.")

    def executar(self):
        
        self.solicitar_entrada()
        self.processar_frase()
        self.contar_ocorrencias()



if __name__ == "__main__":
    analise = FraseAnalise()
    analise.executar()
```

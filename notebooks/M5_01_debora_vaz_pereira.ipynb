```python
numeros = []
with open("dados.txt", "r") as file:
    for linha in file:
        palavras = linha.strip().split(",") 
        for palavra in palavras:
            palavra = palavra.strip()
            try:
                numero = float(palavra)
                numeros.append(numero)
            except ValueError:
                pass

if numeros:
    media = sum(numeros) / len(numeros)
    print(f"média das idades: {media:.2f}")
else:
    print("nenhuma idade encontrada")
```

```python
def adicionar_registro(nome, idade, email):
    #contagem das letras no total e se for menor que 5 é inválido 
    if len(nome) <= 5:
        print("Erro: O nome deve ter mais de 5 caracteres.")
        return

    #se não tiver digitos em idade ou os numeros inteiros forem menor que zero é negativo
    if not idade.isdigit() or int(idade) <= 0:
        print("Erro: A idade deve ser um número positivo.")
        return

    #se @ não estiver dentro de email ou . no email , o email é inválido 
    if "@" not in email or "." not in email:
        print("Erro: O email não é válido.")
        return

    with open("dados.txt", "a") as file:  
        file.write(f"{nome},{idade},{email}\n")
    print("adicionado com sucesso")

adicionar_registro("Pedroalves", "28", "pedroalves@example.com")
adicionar_registro("Anapereira", "20", "anaexample.com")
adicionar_registro("Jo", "30", "jo@example.com")
adicionar_registro("Carlos", "-5", "carlos@example.com")  
```
